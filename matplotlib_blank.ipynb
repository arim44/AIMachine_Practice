* plt.plot([x데이터], y데이터)

* '[ ]'(대괄호)안에 있는 내용은 생략가능, x데이터가 생략될 경우에 x데이터는 0부터 y데이터의 갯수만큼 1씩증가

### 차트 제목, 축 이름 설정

```
plt.plot(["Seoul","Paris","Seattle"], [30,25,55])
plt.xlabel('City')
plt.ylabel('Response')
plt.title('Experiment Result')
plt.show()
```

### 복수의 데이터사용

```
plt.plot([1,2,3],[1,4,9])
plt.plot([2,3,4],[5,6,7])
plt.xlabel('Sequence')
plt.ylabel('Time(secs)')
plt.legend(['CAT','DOG'])
plt.title('Experiment Result')
plt.show()
```

### plot함수의 매개변수들

* plot(\*args, scalex=True, scaley=True, data=None, \**kwargs)
* 매개변수에 \* 하나인것은 tuple
* 매개변수에 \** 두개인것은 dictionary

#### args

plot([x데이터1], y데이터1, [fmt], [x데이터2], y데이터2, [fmt], [x데이터3], y데이터3, [fmt],.....)

```
plt.plot([1,2,3,4],[3,4,5,6],[2,3,4,5],[6,7,8,9])
plt.show()
```

### subplot
* 두개이상의 차트가 필요할때 사용

```
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(2,1,1)
plt.plot(x1,y1,'o-')
plt.title('')
plt.ylabel('')
plt.show()

plt.subplot()
plt.plot(x2,y2, '.-')

plt.show()
```

```
x = np.arange(1,10)
y1 = x*5
y2 = x*1
y3 = x*0.3
y4 = x*0.2

plt.subplot(1,2,1)
plt.plot(x,y1)
plt.subplot(1,2,2)
plt.plot(x,y2)
plt.show()
```

```
x = np.arange(1,10)
y1 = x*5
y2 = x*1
y3 = x*0.3
y4 = x*0.2

plt.subplot(2,2,1)
plt.plot(x, y1)
plt.subplot(2,2,2)
plt.plot(x, y2)
plt.subplot(2,2,3)
plt.plot(x, y3)
plt.subplot(2,2,4)
plt.plot(x, y4)

plt.show()
```

### Bar chart

```
y = [5, 3, 7, 10, 9, 5, 3.5, 8]
x = range(len(y))

plt.bar(x,y, width=0.7, color='orange')
plt.show()
```

### Pie chart

```
labels = ['Apple', 'Samsumg', 'LG', 'HUAWEI']
sizes = [35, 25, 10, 30]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0.1, 0.1, 0, 0)

plt.title('Pie Chart')
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%.2f%%', shadow=True, startangle=90)
plt.show()
```

- size 는 다 합치면 100
- explode
- autopct 숫자 몇자리까지 보여줄지

### scatter (산포도)

```
np.random.seed(1)
X = np.random.normal(0, 1, 100)
Y = np.random.normal(0, 1, 100)
plt.scatter(X,Y)
plt.show()
```

- 분포 확인
- 인공지능할 때는 seed 꼭 고정해놓고 써야함
